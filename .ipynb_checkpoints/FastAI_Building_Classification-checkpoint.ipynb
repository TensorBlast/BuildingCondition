```python
from fastai import *
from fastai.vision.all import *
from fastbook import *
```

```python
path = Path('/home/moot/Monocles/Building Condition')
```

```python
import re

def label_func(a):
    #pat = r'.*(_)(.*)\.jpg'
    pat = r'([^_]+)\.jpg$'
    re.compile(pat)
    return re.search(pat, a.name).group(1)
```

```python
dblock = DataBlock(blocks=(ImageBlock, CategoryBlock),
                   get_items= get_image_files,
                   get_y = label_func,
                   splitter= RandomSplitter(),
                   item_tfms = Resize(224),
                   batch_tfms = aug_transforms())
```

```python
dls = dblock.dataloaders(path)
```

```python
dls.valid.show_batch(max_n=20, nrows=5)
```

```python
learner = vision_learner(dls, xresnet18, metrics=[error_rate, accuracy])
```

```python
learner.fine_tune(5)
```

```python
learner2 = vision_learner(dls, xresnet101, metrics=[accuracy])
```

```python
learner2.fine_tune(5)
```
